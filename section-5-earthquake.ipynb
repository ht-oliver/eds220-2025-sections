```python
# Split time column into date and time
raw_eqk[['date','time_1']] = raw_eqk.time.str.split(" ",expand=True)

# Remove original time column and new time column
raw_eqk.drop(['time', 'time_1'], axis = 1, inplace = True)
```

```python
# Check time range
raw_eqk.date.unique()
```

```python
# Change date to datetime
raw_eqk.date = pd.to_datetime(raw_eqk.date)
```

```python
himag_eqk = raw_eqk[(raw_eqk['type'] == 'earthquake') & (raw_eqk['mag'] >= 6.5)]
```
